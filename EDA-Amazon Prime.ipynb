# **Project Summary -**

This data set was created to list all shows available on Amazon Prime streaming, and analyze the data to find interesting facts. This dataset has data available in the United States.

This dataset has two files containing the titles (titles.csv) and the cast (credits.csv) for the title.

This dataset contains +9k unique titles on Amazon Prime with 15 columns containing their information, including:

**title.csv**
1) id: The title ID on JustWatch.
2) title: The name of the title.
3) show_type: TV show or movie.
4) description: A brief description.
5) release_year: The release year.
6) age_certification: The age certification.
7) runtime: The length of the episode (SHOW) or movie.
8) genres: A list of genres.
9) production_countries: A list of countries that produced the title.
10) seasons: Number of seasons if it's a SHOW.
11) imdb_id: The title ID on IMDB.
12) imdb_score: Score on IMDB.
13) imdb_votes: Votes on IMDB.
14) tmdb_popularity: Popularity on TMDB.
15) tmdb_score: Score on TMDB.
16) And over +124k credits of actors and directors on Amazon Prime titles with 5 columns containing their information:

**credits.csv**
1) person_ID: The person ID on JustWatch.
2) id: The title ID on JustWatch.
3) name: The actor or director's name.
4) character_name: The character name.
5) role: ACTOR or DIRECTOR.

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.

# title.csv file
# First step: we should check for no.of duplicate values so that we could drop them
print("Duplicate values:",title_csv_duplicate)   # the variable title_csv_duplicate is taken from the Duplicate Values cell when were finding duplicates

# drop duplicates
drop_title_duplicates=title_csv.drop_duplicates(inplace=True)  # inplace is used to make changes permanently

# to verify
print("Duplicate values after dropping:",title_csv.duplicated().sum())
```

```python
# Fill the NULL/missing values

# As the dataset is large we should not drop values because we may loose many data
# so it's better to use fillna

# title.csv file
print(title_csv.isnull().sum())

# we can see that columns: description, age_certification, seasons, imdb_id, imdb_score, imdb_votes, tmdb_popularity, tmdb_score has many NULL values
```

```python
# filling/dropping null values

title_csv['description'].fillna('Description Not Available',inplace=True)

title_csv['age_certification'].fillna('Not Rated',inplace=True)

title_csv['seasons'].fillna(0,inplace=True)

title_csv.drop('imdb_id',axis=1,inplace=True)

title_csv['imdb_score'].fillna(title_csv['imdb_score'].median(),inplace=True)

title_csv['imdb_votes'].fillna(title_csv['imdb_votes'].median(),inplace=True)

title_csv['tmdb_popularity'].fillna(title_csv['tmdb_popularity'].median(),inplace=True)

title_csv['tmdb_score'].fillna(title_csv['tmdb_score'].median(),inplace=True)

```

```python
# credits.csv

# First check for duplicate values and then drop those values

print("Duplicate values:",credit_csv_duplicate)  # the variable credit_csv_duplicate is taken from the Duplicate Values cell when were finding duplicates

# drop duplicates
drop_credit_duplicates=credits_csv.drop_duplicates(inplace=True)

# to verify
print("Duplicate values after dropping:",credits_csv.duplicated().sum())

```

```python
# Fill the NULL/missing values

# As the dataset is large we should not drop values because we may loose many data
# so it's better to use fillna

# credits.csv file
print(credits_csv.isnull().sum())

# we can see that columns: character has NULL/missing values
```

```python
# filling the NULL/missing values in character column of credits.csv file

credits_csv['character'].fillna('Not Applicable', inplace=True)

# Filled the NaN fields with "Not Applicable" because if we drop those rows than we may loose many data
```
